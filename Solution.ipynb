# 7번 문제 풀이
Alice의 isogeny $\phi_A$를 구성하는 과정에서, `Isogeny Test.ipynb`에 적었듯 2-isogeny를 $e_2$개 구해 합성한다. 여기서 주목할 점은 $\phi_A$가 다음 식처럼 표현될 때,

$$
\phi_A = \phi^{(e_2 - 1)}_A \circ \dots \circ \phi^{(0)}_A
$$

Dual isogney $\hat{\phi}_A$는 다음과 같이 표현할 수 있고,

$$
\hat{\phi}_A = \hat{\phi}^{(0)}_A \circ \dots \circ \hat{\phi}^{(e_2 - 1)}_A
$$

2-isogeny인 $\hat{\phi}^{(i)}_A$의 kernel은 $\langle R, S \rangle = E[2^{e_2}]$인 어떤 $S \in E[2^{e_2}]$에 대해 다음이 성립한다.

$$
\ker \hat{\phi}^{(i)}_A = \left\langle \left( \phi^{(i)}_A \circ \dots \circ \phi^{(0)}_A \right)\left( [2^{e_2 - 1}]S \right) \right\rangle
$$

SIKE 표준이나 우리 문제와 같은 경우에 $S = Q_A$로 둘 수 있다. 그런데, $[2^{e_2 - 1}]S = [2^{e_2 - 1}]Q_A$를 계산하면 다음과 같다.

```python
2^(e_2 - 1) * Q_A
```

커널이 $\langle (0, 0) \rangle$이 아닌 2-isogeny $\phi$에 대해 $\phi(0, 0) = (0, 0)$이므로, 2-isogeny인 $\hat{\phi}^{(i)}_A$의 kernel은 $\langle (0, 0) \rangle$이다. 한편 $\phi^{(e_2 - 1)}_A: \dots \to E_A$이므로 $\hat{\phi}^{(e_2 - 1)}_A$의 정의역은 $E_A$이다. $\hat{\phi}^{(e_2 - 1)}_A$의 공역은 $\hat{\phi}^{(e_2 - 2)}_A$의 정의역... 이므로 반복하면 dual isogeny를 구할 수 있다... 라고 믿고 싶지만 되지 않는데.

```python
E_tmp = E_A
dual_components = []
for j in range(e_2):
    print(j)
    phi = E_tmp.isogeny(E_tmp(0, 0))
    E2 = phi.codomain()
    a2 = E2.a2()
    E3 = EllipticCurve(k, [0, a2 + 6, 0, 4*(a2 + 2), 0])
    E4 = E3.quadratic_twist(sqrt(1 / k(16 * E3.a4())))
    psi = E2.isomorphism_to(E4)
    dual_components.append(psi * phi)
    E_tmp = dual_components[-1].codomain()
    print(dual_components[-1])
    print()
```
